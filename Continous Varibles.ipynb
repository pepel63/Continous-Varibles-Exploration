```python
import pandas as pd

# Read CSV file into a DataFrame
df = pd.read_csv('pbc.csv')

# Extract the 'chol' column
chol_values = df['chol']

# Sort the values in ascending order is not necessary for quantile calculation
# Pandas can compute quantiles directly from the Series

# Calculate quartiles
Q1 = chol_values.quantile(0.25)
Q2 = chol_values.quantile(0.5)  # Median
Q3 = chol_values.quantile(0.75)

print("Q1 (25th percentile):", Q1)
print("Q2 (Median, 50th percentile):", Q2)
print("Q3 (75th percentile):", Q3)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('pbc.csv')
plt.figure(figsize=(8,6))
df.boxplot(column='chol')
plt.title('Boxplot of Cholesterol (chol)')
plt.ylabel('Cholestrol')
plt.show()
```

```python
sorted_chol = sorted(df['chol'])

Q1 = pd.Series (sorted_chol).quantile(0.25)

Q3 = pd.Series(sorted_chol).quantile(0.75)

# Calculate interquartile range (IQR)

IQR=Q3 - Q1

#Calculate upper and lower fences

upper_fence = Q3 + (1.5 * IQR)

lower_fence = Q1 - (1.5 * IQR)

#Identify outliers

outliers = df [(df ['chol'] > upper_fence) | (df['chol'] < lower_fence)]

print("IQR:", IQR)

print("Upper Fence:", upper_fence)

print("Lower Fence:", lower_fence)

print("Outliers:")

print(outliers)
```
